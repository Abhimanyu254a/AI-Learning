# LangChain Version v1

```python
import os
from dotenv import load_dotenv

load_dotenv()

os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
# this is how you make an agent
from langchain.agents import create_agent

def get_weather(city:str) -> str:
    # this doc string is also consume by the agent
    """
    Get the Weather for a city, this function is a tool which is used by the agent to take the 
    context and also use this function to find the answer.
    """
    return f"The weather is {city} is sunny." 

     
agent=create_agent(
    model="gpt-5",
    tools=[],
    system_prompt="You are a helpful assistant"
)

agent
```

```python
# running the Agent
agent.invoke({
    "messages":[{
        "role":"user",
        "content":"What is the weather like in Delhi?"
    }]
})
```

```python
# running the agent in 2nd methods
agent.invoke({
    'messages':"what is the weather in the Delhi?"
    })
```
