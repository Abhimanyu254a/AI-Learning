## Message

Message are the fundamental unit of context for models in LangChain. They present the input and output of models, carrying both the content and metadata needed to represent the state of conversation when interacting with an LLM. Messages are objects that contains:
- Role - Identifies the message(eg system, user)
- Content - Respresents the actual content of the message(like text, images, audio, documents, etc)
- Metadata - Optional fields such as response information, message IDs, and token useage

LangChain provides a standard message type that works across all model providers, ensuring consistent behavior regardless of the model being called.

```python
# import the Models
import os
from langchain.chat_models import init_chat_model

os.environ['API_KEY'] = os.getenv('API_KEY')

model = init_chat_model("Model_Name")
```

### Message Prompt

Alternatively, you can pass in a list of messages to the model by providing a list of message objects.

Message Types
- System message - Tells the model how to behave and provide context for interactions
- Human message - Respresents user input and interactions with the model
- AI message - Responses generated by the model, including text content, tool calls, and metadata
- Tools Message - Represent the outputs of tools calls

### System Message

A SystemMessage respesent an initial set of instructions that primes the model's behavior. You can use a system message to set the tone, define the model's role, and establish guidelines for responses

### Human Message

A HumanMessage respresents user input and interactions. They can contain text, images, audio, files,and other amount of multimodel contents.

### AI Message

An AIMessage represents the output of a model invocation. They can include multimodel data, tools calls, and provider-specific metadata that you can later access

### Tool Message

For Models that support tool calling, AI message can contain tools calls. Tool message are used to pass the results of a single tool execution back to the model


```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage
```

```python

messages=[
    SystemMessage("You are a poetry expert"),
    HumanMessage("Write a poem on AI"),
]
response=model.invoke(messages)
response.content
```

```python
system_mes = SystemMessage("You are a helpful coding assistant")
messages=[
    system_mes,
    HumanMessage("How do I create a REST API?"),
]
response=model.invoke(messages)
print(response.content)
```

```python
## Detailed info to the LLM through System Message

from langchain.messages import SystemMessage, HumanMessage

System_msg = SystemMessage("""
    You are a senior Python developer with expertise in the web framework.
    Always provide code examples and explain your resoning. 
    be concise but thorough in your explanations.
""")

message =[
    system_mes,
    HumanMessage("How do I create a REST API?")
]

response = model.invoke(message)
print(response.content) 
```

```python
# Message Metadata

human_mes = HumanMessage(
    content="hello",
    name="Abhimanyu",
    id="msg_254"
)
```

```python
response = model.invoke([
    human_mes
])
response
```

```python
# create an AI Message manually(eg for conversation history)
ai_msg = AIMessage("I'd be happy to help you with that question!")

#Add to the Conversation history

message = [
    SystemMessage("You are a helpful Assistant"),
    HumanMessage("Can you Help me?"),
    ai_msg,
    HumanMessage("Great What's 2+2")
]

response = model.invoke(message)
print(response.content)
```

```python
# To see the Meta data
response.usage_metadata
```

```python
# Tools

from langchain.messages import AIMessage
from langchain.messages import ToolMessage

## After a model makes a tool Call
## Here, we demonstrate manually creating the messages for brevity

ai_message = AIMessage(
    content=[],
    tool_calls=[{
        "name":"get_weather",
        "args":{"location":"Delhi"},
        "id":"call_123"
    }]
)

## Execute Tool and create result message
weather_result = "Hot, 45C"
tool_message = ToolMessage(
    content=weather_result,
    tool_call_id="call_123" #Must match the Call ID
)

# continue  Conversation

message = [
    HumanMessage("What's the weather in Delhi?"),
    ai_message, # model's tool cell
    tool_message # Tool execution result
]

response = model.invoke(message) # model processes the result
```
