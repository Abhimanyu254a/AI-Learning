## Tools

Models can request to call tools that performs tasks such as fetching data from a database, searching the web, or running code. Tools are pairing of:-

- A Schema, including the name of the tool, a description, and/or argument definitaions 
- A functin or coroutine to execute

```python
# import a Model and then you all the things that I am going use it
```

```python
from langchain.tools import tool

@tool
def get_weather(location:str) -> str:
    """Get the weather at a location"""
    return f"It's sunny in {location}"

model_with_tools = model.bind_tools([get_weather])
```

```python
response = model_with_tools.invoke("What's the weather like in Delhi?")
print(response)

for tool_call in response.tool_calls:
    print(f"Tools {tool_call['name']}")
    print(f"Args: {tool_call['args']}")
```
